```python
from bokeh.plotting import show, Figure, output_notebook
from bokeh.models import Label, ColumnDataSource, HoverTool
output_notebook()
```

```python
import pandas as pd
```

```python
import tabula
```

```python
df = tabula.read_pdf('DATABASE_Sep2020_web.pdf', pages = 5, lattice = True)
```

```python
data = df[0].iloc[4:].reset_index(drop=True)
data.Year = data.Year.apply(lambda x: int(x.split("-")[0]))
data.Total = data.Total/1000
```

```python
url = 'file:///home/robin/troll_graphs/coffeebean.png'
data['url'] = url
data['w'] = 1.3
data['h'] = 8.3
data.head()
```

```python
data.info()
```

```python
source = ColumnDataSource(data[["Year","Total",'url','h','w']])
plot = Figure(
    title = "Coffee Production In 1000 Mt Tonne",
    x_axis_label = "Year", y_axis_label = "Coffee Production (1000 Mt Tonnes)", x_range= (1989, 2030)
)

plot.xaxis.ticker = list(range(1989,2030,5))
plot.circle(
    x = 'Year', y = 'Total', visible = True, name = 'glip', source =source, size = 15
)
plot.image_url(
    url = 'url', x = 'Year', y = 'Total',
    w='w', h='h', source = source, anchor = 'center'
)

ht = HoverTool(names=['glip'])
ht.tooltips = [('Year','@Year'),('Quantity','@Total')]
plot.add_tools(ht)
show(plot)
```
